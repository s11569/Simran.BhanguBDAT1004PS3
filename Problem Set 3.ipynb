#### Step 1 -  Import the necessary libraries

#### Step 2 - Import the dataset from this address https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/Wind_Stats/wind.data

```python
wind_stats = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/Wind_Stats/wind.data", delim_whitespace=True)
wind_stats.head(10)
```

```python
wind_stats.info()
```

#### Step 3 - Assign it to a variable called data and replace the first 3 columns by a proper datetime index

```python
def get_month_or_day(val):
    if val < 10:
        return(str(0)+str(val))
    else:
        return(str(val))

wind_stats["date"] = wind_stats[["Yr", "Mo", "Dy"]].apply(lambda z: str(z["Yr"])+"-"+get_month_or_day(z["Mo"])+"-"+get_month_or_day(z["Dy"]), axis=1 )
wind_stats["date"] = pd.to_datetime(wind_stats["date"], format="%y-%m-%d")
wind_stats.head()
```

```python
wind_stats["date"].dtypes
```

#### Step 4 - Year 2061? Do we really have data from this year? Create a function to fix it and apply it.

```python
wind_stats["date"] = wind_stats[["Yr", "Mo", "Dy"]].apply(lambda z: "19"+str(z["Yr"])+"-"+get_month_or_day(z["Mo"])+"-"+get_month_or_day(z["Dy"]), axis=1 )
wind_stats["date"] = pd.to_datetime(wind_stats["date"])
wind_stats.head()
```

```python
wind_stats.drop(["Yr", "Mo", "Dy"], axis=1, inplace=True)
```

#### Step 5 - Set the right dates as the index. Pay attention at the data type, it should be datetime64[ns]

```python
wind_stats.set_index("date", inplace=True)
wind_stats.head()
```

#### Step 6 - Compute how many values are missing for each location over the entire record.They should be ignored in all calculations below.

####  Column Wise null Values in record

```python
wind_stats.isnull().sum()
```

####  Row wise values in record

```python
wind_stats[wind_stats.isnull().any(axis=1)]
```

#### Step 7 - Compute how many non-missing values there are in total.

```python
print(wind_stats.info())
print(wind_stats.notnull().sum())
```

#### Step 8 - Calculate the mean windspeeds of the windspeeds over all the locations and all the times

##### A single number for the entire dataset.

```python
wind_stats.mean(axis=1).mean(axis=0)
```

#### Step 9 - Create a DataFrame called loc_stats and calculate the min, max and mean windspeeds and standard deviations of the windspeeds at each location over all the days

##### A different set of numbers for each location.

```python
loc_stats = pd.DataFrame()
loc_stats['min'] = wind_stats.min(axis=0)
loc_stats['max'] = wind_stats.max(axis=0)
loc_stats['mean'] = wind_stats.mean(axis=0)
loc_stats['std'] = wind_stats.std(axis=0)
```

```python
loc_stats.head(10)
```

#### Step 10 - Create a DataFrame called day_stats and calculate the min, max and mean windspeed and standard deviations of the windspeeds across all the locations at each day.

##### A different set of numbers for each day

```python
day_stats = pd.DataFrame()
day_stats['min'] = wind_stats.min(axis=1)
day_stats['max'] = wind_stats.max(axis=1)
day_stats['mean'] = wind_stats.mean(axis=1)
day_stats['std'] = wind_stats.std(axis=1)
```

```python
day_stats.head(10)
```

#### Step 11 - Find the average windspeed in January for each location.

##### Treat January 1961 and January 1962 both as January.

```python
wind_stats = wind_stats.reset_index() 
Jan_stats = wind_stats[wind_stats['date'].dt.month == 1]
Jan_stats = Jan_stats.mean(axis = 0)
```

```python
Jan_stats
```

#### Step 12 - Downsample the record to a yearly frequency for each location

```python
wind_stats.groupby(wind_stats['date'].dt.year).sum()
```

#### Step 13 - Downsample the record to a monthly frequency for each location

```python
data_mnth = wind_stats
data_mnth.groupby([data_mnth['date'].dt.year.rename('year'),data_mnth['date'].dt.month.rename('month')]).sum()
```

#### Step 14 - Downsample the record to a weekly frequency for each location.

```python
data_week = wind_stats
data_week.groupby([data_week['date'].dt.year.rename('year'),data_week['date'].dt.month.rename('month'), data_week['date'].dt.week.rename('week')]).sum()
```

#### Step 15 - Calculate the min, max and mean windspeeds and standard deviations of the windspeeds across all locations for each week (assume that the first week starts on January 2 1961) for the first 52 weeks.

```python
calculate_data = data_week.groupby([data_week['date'].dt.year.rename('Year'),data_week['date'].dt.week.rename('Week')]).sum()

calculate_mean = pd.Series(calculate_data.head(52).mean(axis=1))
calculate_min = pd.Series(calculate_data.head(52).min(axis=1))
calculate_max = pd.Series(calculate_data.head(52).max(axis=1))
calculate_std = pd.Series(calculate_data.head(52).std(axis=1))
```

```python
calculate_stats = pd.DataFrame({'Min':calculate_min, 'Max':calculate_max, 'Mean':calculate_mean, 'Std Deviation':calculate_std})
calculate_stats
```
